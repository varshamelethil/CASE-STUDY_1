#### 3. Plot a bar graph of male vs female buyers participated in the sales 

```python
count=car_data['Buyer Gender'].value_counts()
count
```

```python
car_data['Buyer Gender'].value_counts().plot.bar(width=0.5, color=['red','green'])
plt.xlabel('Buyer')
plt.ylabel('Count')
plt.title("Male vs Female buyers")
plt.grid()
plt.show()
```

Data insights:

* The data contain total 10000 buyers
